** import the dataset**


```python
df=pd.read_csv(r"walmart-10k-sales-datasets\Walmart.csv",encoding_errors="ignore")
#to check the number of rows and columns in the dataset
df.shape
```

```python
# to see the first 5 rows 
df.head()
```

```python
# gives you a quick statistical summary of the numerical columns in your DataFrame.
df.describe()
```

```python
#gives the datatype of each column and non-null counts
df.info()
```

```python
#to count all the duplicate rows in the dataset
df.duplicated().sum()
```

```python
#command to remove the duplicates 
df.drop_duplicates(inplace=True)
```

```python
# after removing the duplicate values the no of rows reduces
df.shape
```

```python
#tells which column have how many null values
df.isnull().sum()
```

```python
# to remove the null values we are deleting the entire row which have nul values
df.dropna(inplace=True)
```

```python
# to change the datatype of the column 
df['unit_price'].astype('float')
```

```python
#to remove the $ sign from the unit price column and change the datatype to float
df['unit_price']=df['unit_price'].str.replace('$','').astype('float')
```

```python
#to make a new column using price and quantity column
df['total_price']=df['unit_price'] * df['quantity']
```
